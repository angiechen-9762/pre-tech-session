### 11. Which release year saw the lowest average user price?

```python
#1
data_year = data.groupby('ReleaseYear').mean()
data_year.reset_index(inplace=True)
data_year[data_year.LowUserPrice == data_year.LowUserPrice.min()][['ReleaseYear','LowUserPrice']]
```

```python
#2 - sort_values()
data.groupby('ReleaseYear')['LowUserPrice'].mean().sort_values()
```

```python
#3 - groupby(as_index=False), sort_values(ignore_index=True): indexing after ordering
data.groupby(["ReleaseYear"], as_index = False)["LowUserPrice"].mean().\
    sort_values('LowUserPrice', ascending=True, ignore_index=True)
```

### 12. Reshape/ Pivot the data to create a dataframe that has SKUs in one column, all continuous numeric variables (StrengthFactor, PriceReg,ItemCount, LowUserPrice, LowNetPrice) as values in one column and all their respective numerical values in a 3rd column (Hint: this dataframe should have only 3 columns in total- SKU, Variable, Value)

```python
#1 - melt
data_melt = data.melt(id_vars='SKU_number',\
            value_vars=['StrengthFactor','PriceReg','ItemCount','LowUserPrice','LowNetPrice'],\
            var_name='Variable', value_name='Value')
data_melt.rename(columns={'SKU_number':'SKU'})
data_melt.set_index('SKU_number').sort_index()
```

```python
#2 - stack
data_reshape = data[['SKU_number','StrengthFactor','PriceReg','ItemCount','LowUserPrice','LowNetPrice']].set_index('SKU_number').stack(level=0)
data_reshape = pd.DataFrame(data_reshape,columns = ['Values'])
data_reshape
```

### 13. Create a chart/plot to show average lownetprice trend by release year for last 10 years

```python
#1
data_10yr = data_year.tail(10)
data_10yr.plot('ReleaseYear', 'LowNetPrice')
```

```python
#2 - .dropna
data=data.dropna(axis=0) # drop nan for drawing
data_10yr = data.groupby(["ReleaseYear"], as_index = False)["LowNetPrice"].mean().\
    sort_values('ReleaseYear', ascending=True, ignore_index=True).tail(10)
data_10yr.plot('ReleaseYear', 'LowNetPrice')
plt.title('Average LowNetPrice by Year')
plt.ylabel('Average LowNetPrice')
plt.xlabel('Years')

plt.show()
```
