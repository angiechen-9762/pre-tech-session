```python
data_year = data.groupby('ReleaseYear').mean()
data_year.reset_index(inplace=True)
data_year[data_year.LowUserPrice == data_year.LowUserPrice.min()].ReleaseYear
```

```python
data_melt = data.melt(id_vars='SKU_number',\
            value_vars=['StrengthFactor','PriceReg','ItemCount','LowUserPrice','LowNetPrice'],\
            var_name='Variable', value_name='Value')
data_melt.rename(columns={'SKU_number':'SKU'})
```
